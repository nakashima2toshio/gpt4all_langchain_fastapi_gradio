# テキスト生成

```python
# メッセージリストの準備
messages = [
    {
        "role": "system",
        "content": "あなたはQAチャットボットです。"
    },
    {
        "role": "user",
        "content": "日本一高い山は？",
    }
]
```

```python
# テキスト生成
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=messages,
    temperature=0.2,
    max_tokens=500
)
show_json(response)
```

```python
# メッセージコンテンツの取得
print(response.choices[0].message.content)
```

# ストリーミング

```python
# メッセージリストの準備
messages = [
    {
        "role": "system",
        "content": "あなたはQAチャットボットです。"
    },
    {
        "role": "user",
        "content": "日本一高い山は？",
    }
]
```

```python
# テキスト生成
stream = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=messages,
    temperature=0.2,
    max_tokens=500,
    stream=True,  # ストリーミングの有効化
)

# ストリーミングの出力
for part in stream:
    print(part.choices[0].delta.content or "")
```

# JSONモード

```python
# メッセージリストの準備
messages = [
    {
        "role": "system",
        "content": "あなたは役に立つアシスタントで、全てJSON形式で出力します。"
    },
    {
        "role": "user",
        "content": "東京の観光名所は？"
    },
]
```

```python
# テキスト生成
response = client.chat.completions.create(
    model="gpt-3.5-turbo-1106",
    response_format={"type":"json_object"},  # JSONモードの有効化
    messages=messages,
    temperature=0.9,
    max_tokens=500
)
print(response.choices[0].message.content)
```

```python
# finish_reasonの確認
print(response.choices[0].finish_reason)
```

# 非同期処理

```python
# Colab用の前処理
import nest_asyncio
nest_asyncio.apply()
```

```python
import asyncio
from openai import AsyncOpenAI

# 非同期用のクライアントの準備
async_client = AsyncOpenAI()
```

```python
# メッセージリストの準備
messages=[
    {
        "role": "user",
        "content": "日本一高い山は？",
    }
]
```

```python
# テキスト生成
async def main() -> None:
    response = await async_client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=0.2,
        max_tokens=500,
    )
    print(response.choices[0].message.content)

# 実行
asyncio.run(main())
```
