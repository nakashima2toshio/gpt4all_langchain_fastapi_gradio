```python
import tiktoken

# エンコーディングの取得
enc = tiktoken.get_encoding("cl100k_base")
```

```python
# エンコードの実行
tokens = enc.encode("Hello World!")
print(len(tokens))
print(tokens)
```

```python
# デコードの実行
print(enc.decode(tokens))
```

```python
# 分割したままでデコードを実行
print(enc.decode_tokens_bytes(tokens))
```

# 日本語と英語のトークン数の比較

```python
# エンコードの実行
tokens = enc.encode("こんにちは、世界！")
print(len(tokens))
print(tokens)
```

```python
# デコードの実行
print(enc.decode(tokens))
```

```python
# 分割したままデコード
def data2str(data):
    try:
        return data.decode('utf-8')
    except UnicodeError:
        return data
print([data2str(data) for data in enc.decode_tokens_bytes(tokens)])
```
