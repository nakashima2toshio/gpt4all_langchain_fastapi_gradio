```python
from openai import OpenAI

# クライアントの準備
client = OpenAI()
```

```python
import json

# JSON出力ヘルパーの準備
def show_json(obj):
    display(json.loads(obj.model_dump_json()))
```

# テキストからの画像生成 - 画像URL

```python
# プロンプトの準備
prompt = "カフェで働いている猫耳メイドがコップの水をこぼしたイラスト"
```

```python
# テキストからの画像生成
response = client.images.generate(
    model="dall-e-3",
    prompt=prompt,
    size="1024x1024",
    quality="standard",
    n=1,
)
show_json(response)
```

```python
# 画像URLの取得
print(response.data[0].url)
```

# テキストからの画像生成 - 画像データ

```python
# プロンプトの準備
prompt = "カフェで働いている猫耳メイドがコップの水をこぼしたイラスト"
```

```python
# テキストからの画像生成
response = client.images.generate(
    model="dall-e-3",
    prompt=prompt,
    size="1024x1024",
    quality="standard",
    n=1,
    response_format="b64_json",
)
show_json(response)
```

```python
# 画像データの取得
print(response.data[0].b64_json)
```

```python
import base64
from PIL import Image
from io import BytesIO

# 画像データ (base64) を画像ファイルに保存
image_data = base64.b64decode(response.data[0].b64_json)
image_stream = BytesIO(image_data)
image = Image.open(image_stream)
image.save("image.png")
```

# 既存の画像の編集 (DALL-E 2 のみ)

```python
# 画像とマスクの準備 (画像とマスクをカレントフォルダに配置)
image = open("image.png", "rb")
mask = open("mask.png", "rb")
```

```python
# プロンプトの準備
prompt = "many apples in cardboard box"
```

```python
# 既存の画像の編集
response = client.images.edit(
    model="dall-e-2",
    image=image,
    mask=mask,
    prompt=prompt,
    n=1,
    size="512x512"
)
show_json(response)
```

```python
# 画像URLの取得
print(response.data[0].url)
```

# 既存の画像のバリエーション生成（DALL-E 2 のみ）

```python
# 画像の準備 (画像をカレントフォルダに配置)
image = open("image.png", "rb")
```

```python
# 既存の画像のバリエーション生成
response = client.images.create_variation(
    model="dall-e-2",
    image=image,
    n=1,
    size="512x512"
)
show_json(response)
```

```python
# 画像URLの取得
print(response.data[0].url)
```
